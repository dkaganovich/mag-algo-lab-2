# 3. Случайный граф Кронекера

## Конфигурация

Код генерации ребер графа был взят из [референс кода Graph500](http://www.graph500.org/referencecode). Для подсчета различных статистик использовалась библиотека [BGL](http://www.boost.org/doc/libs/1_62_0/libs/graph/doc/index.html). Всего было сгенерировано **1000** графов с базовой матрицей $\left[ {\begin{array}{cc} \alpha & \beta\\\beta & \gamma\\ \end{array} } \right]$=$\left[ {\begin{array}{cc} 0.57 & 0.19\\0.19 & 0.05\\ \end{array} } \right]$, числом вершин равным $2^{12}=4096$ и *edge_factor*-ом равным 16.

## Статистики графа

### Small World Pattern

На рисунке ниже приведены подсчитанные диаметр и среднее расстояние для сгенерированных графов: 

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def ax_init(ax, title, xlbl, ylbl, xlim, xtick, ylim, ytick):
    ax.set_title(title)
    ax.set_xlabel(xlbl)
    ax.set_xlim(xlim)
    ax.set_xticks(np.arange(xlim[0], xlim[1], xtick))
    ax.set_ylabel(ylbl)
    ax.set_ylim(ylim)
    ax.set_yticks(np.arange(ylim[0], ylim[1], ytick))
    ax.grid(True)
    
def plot_small_world_file(filename, xlim=[0, 1000], xtick=50, ylim=[0, 10], ytick=1):
    trial, diam, dist = [], [], []
    cnt = 0
    for line in open(filename, "r"):
            tkns = line.split(",")
            if len(tkns) != 2:
                raise ValueError("Invalid file format")
            trial.append(cnt)
            diam.append(int(tkns[0]))
            dist.append(float(tkns[1]))
            cnt += 1;
    fig, axarr = plt.subplots(1, 1, figsize=(16, 8))
    ax_init(axarr, "Small World Pattern", "Trial Number", "Number of Edges", xlim, xtick, ylim, ytick);
    axarr.scatter(trial, diam, c="red", label="Diameter", edgecolors="none")
    axarr.scatter(trial, dist, c="blue", label="Mean Distance", edgecolors="none")
    axarr.legend(loc='best', frameon=False)
    plt.tight_layout()
    plt.show()
    
plot_small_world_file("--skg_small_world_5700_1900.csv")
```

Здесь можно отметить небольшой диаметр графа, что характерно для реальных web-графов. При удачном выборе базовой матрицы ($\beta+\gamma > 1$), диаметр стабилизируется по мере увеличения числа вершин (см. [kronecker3.pdf](kronecker3.pdf)), что также является свойством реальных web-графов.

### <a name="nd"></a>\#(n,d)

Ниже представлен общий вид распределения степеней по вершинам графа: 

```python
def plot_rank_hist_file(filename, xlim=[0, 1000], xtick=25, ylim=[0, 0.3], ytick=0.1):
    rank, perc = [], []
    for line in open(filename, "r"):
            tkns = line.split(",")
            if len(tkns) != 2:
                raise ValueError("Invalid file format")
            rank.append(int(tkns[0]))
            perc.append(float(tkns[1]))

    fig, axarr = plt.subplots(1, 1, figsize=(16, 8))
    ax_init(axarr, "#(n,d)", "Rank", "Vertex Perc.", xlim, xtick, ylim, ytick);
    axarr.scatter(rank, perc, c="red", label="Vertex Perc.", edgecolors="none")
    axarr.legend(loc='best', frameon=False)
    plt.tight_layout()
    plt.show()
plot_rank_hist_file("--skg_rank_hist_5700_1900.csv")
```

Здесь также можно отметить паттерн, наблюдаемый у всех web-графов: степенной закон распределения степеней вершин с "тяжелым хвостом". Т.е. в графе присутствует относительно небольшое число вершин с высокими степенями - хабов, и много вершин с небольшими степенями, которые связаны через хабов. 

### Компоненты связности

На рисунке ниже можно увидеть число комонент связности в сгенерированных графах:

```python
def plot_num_comps_file(filename, xlim=[0, 1000], xtick=100, ylim=[0, 1000], ytick=100):
    trial, ncomps = [], []
    cnt = 0
    for line in open(filename, "r"):
            tkns = line.split(",")
            if len(tkns) != 1:
                raise ValueError("Invalid file format")
            trial.append(cnt)
            ncomps.append(int(tkns[0]))
            cnt += 1
    fig, axarr = plt.subplots(1, 1, figsize=(16, 8))
    ax_init(axarr, "Connected Components", "Trial", "Number of Components", xlim, xtick, ylim, ytick);
    axarr.scatter(trial, ncomps, c="red", label="Number of Components", edgecolors="none")
    axarr.legend(loc='best', frameon=False)
    plt.tight_layout()
    plt.show()
plot_num_comps_file("--skg_num_comps_5700_1900.csv")
```

Все сгенерированные графы получились несвязными. Большое число компонент связности объясняется высоким процентом изолированных вершин ([см. пункт выше](#nd)). Но, снова, в [kronecker3.pdf](kronecker3.pdf) показано, что при определенном выборе коэффициентов базовой матрицы ($\beta+\gamma > 1$), граф можно сделать связным. Таким образом данную модель можно адаптировать под большое число реальных web-графов.

# 4. Граф Боллобаша – Риордана

## Конфигурация

Функция генерации ребер графа находится в исходнике [brg.cc](brg/brg.cc). Статистики считались с помощью [BGL](http://www.boost.org/doc/libs/1_62_0/libs/graph/doc/index.html). Было сгенерировано **100** графов с параметрами N=$2^{12}$ и M=16. Ниже приведены графики аналогичные вышеуказанным.

## Статистики графа

### Small World Pattern

```python
plot_small_world_file("--brg_small_world.csv", xlim=[0, 100], xtick=10, ylim=[0, 10], ytick=1)
```

Отметим константное значение диаметра графа. Здесь, в отличие от графа Кронекера, мы получили это свойство "автоматически". Кроме того, полученное значение диаметра графа полностью согласуется с теоретически предсказанным: $\frac{ln(N)}{ln(ln(N))}=[N=4096]=3.92645\approx4$ 

### \#(n,d)

```python
rank, perc = [], []
for line in open("--brg_rank_hist.csv", "r"):
        tkns = line.split(",")
        if len(tkns) != 2:
            raise ValueError("Invalid file format")
        rank.append(int(tkns[0]))
        perc.append(float(tkns[1]))
fig, axarr = plt.subplots(1, 1, figsize=(16, 8))
ax_init(axarr, "#(n,d)", "Rank", "Vertex Perc.", [0, 1000], 25, [0, 0.3], 0.1);
axarr.scatter(rank, perc, c="red", label="Vertex Perc.", edgecolors="none")
axarr.scatter(rank, [546./r**3 for r in rank], c="blue", label="Vertex Perc. Theor. c=546", edgecolors="none")
axarr.legend(loc='best', frameon=False)
plt.tight_layout()
plt.show()
```

Теория говорит, что по мере увеличения числа вершин в графе наше распределение $\#(n,d)\approx\frac{c}{d^{3}}$. В случае выше выборочно была подсчитана константа $c=546$. Видно, что распределения действительно совпали.

### Компоненты связности

```python
plot_num_comps_file("--brg_num_comps.csv", xlim=[0, 100], xtick=10, ylim=[0, 10], ytick=1)
```

В случае с моделью Боллобаша-Риордана все сгенерированные графы оказались связными.
